```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo de excel sin indice
df = pd.read_excel('Datos de Facturación.xlsx')
df.info
```

```python
#Se utiliza la comparación and con los rangos 1000 y 2000
filtro1000a2000 = df[(df['CVE_CLPV'] > 1000) & (df['CVE_CLPV'] < 2000)]
filtro1000a2000
filtro1000a2000.to_csv('actividad2.2_filtro1.csv')
```

```python
#Se utiliza la comparación or con los valores 4 y 5
filtroNOT = df[~((df["CVE_VEND"] == 4) | (df["CVE_VEND"] == 5))]
filtroNOT
filtroNOT.to_csv('actividad2.2_filtro2.csv')
```

```python
#Se convierte la columna a tipo fecha y se filtra por la fecha 28/02/2020
filtroDATE = df[df["FECHA_ENT"] == "28/02/2020"]
filtroDATE

```

```python
#Se filtra por los valores de la columna STATUS que sean E o sean menores a 5951.7 utilizando el operador or
filtroCantTot = df[(df["STATUS"] == "E")| (df["CAN_TOT"] < 5951.7)]
filtroCantTot
filtroCantTot.to_csv('actividad2.2_filtro4.csv')
```

```python
#Se filtra por las columnas CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtroOnlyColumns = df.loc[:, ["CVE_DOC", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]
filtroOnlyColumns
filtroOnlyColumns.to_csv('actividad2.2_filtro5.csv')
```

```python
#Se filtra por las filas 7001 a 7099 con todas las columnas
filtroRowRange = df.loc[7001:7099, :]
filtroRowRange
filtroRowRange.to_csv("actividad2.2_filtro6.csv")
```

```python
#Se filtra por los valores de la columna CVE_VEND que sean 1 o 2 y se selecciona la columna FECHAELAB
df1 = pd.read_excel('Datos de Facturación.xlsx', index_col= "CVE_VEND") 
filtroIndex = df1.loc[[1,2], ["FECHAELAB"]]
filtroIndex.to_csv("actividad2.2_filtro7.csv")
```
